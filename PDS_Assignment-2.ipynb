```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import re
import datetime
```

```python
# Load the dataset
df = pd.read_csv('train.csv')
```

```python
df
```

```python
# Identify columns with missing values
columns_with_missing_values = df.columns[df.isnull().any()]
columns_with_missing_values
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```python
# Impute missing values or drop them
for col in columns_with_missing_values:
    if df[col].dtype == 'object':
        df[col].fillna(df[col].mode()[0], inplace=True)  # Replace with mode for categorical data
    else:
        df[col].fillna(df[col].mean(), inplace=True)  # Replace with mean for numerical data
        
# Define a function to remove multiple units from a column using regex
def remove_units(text):
    pattern = re.compile(r'[^\d.]') 
    return re.sub(pattern, '', text)

columns_to_clean = ['Mileage', 'Engine', 'Power', 'New_Price']
for col in columns_to_clean:
    df[col] = df[col].apply(remove_units)
    df[col] = pd.to_numeric(df[col], errors='coerce')
    
df.to_csv('new_train.csv',index = False)
```
